```python
import pandas as pd
import random
```

```python
quarters = ["2017Q1", "2017Q2", "2017Q3", "2017Q4", "2018Q1", "2018Q2", "2018Q3","2018Q4","2019Q1", "2019Q2", "2019Q3", 
            "2019Q4", "2020Q1"]

df = pd.DataFrame()

for quarter in quarters:
    temp = pd.read_csv(f"Data\{quarter}.csv")
    temp = temp.sample(frac=.05, random_state = 43)
    df = df.append(temp)
    print(quarter)
```

```python
airportsDf = pd.read_csv('Data/airportData.csv')
airportsDf.head()
```

```python
airportsDf = airportsDf.rename(columns = {
    'IATA_CODE': 'Origin',
    'AIRPORT': 'Origin_Airport',
    'CITY': 'Origin_City',
    'STATE': 'State',
    'COUNTRY': 'Country',
    'LATITUDE': 'Origin_Lat',
    'LONGITUDE': 'Origin_Long'
})
airportsDf.head()
```

```python
originsDf = airportsDf[[
    'Origin',
    'Origin_Airport',
    'Origin_City',
    'Origin_Lat',
    'Origin_Long'
]]
```

```python
totalDf = pd.merge(joinedDf, originsDf, on='Origin')
```

```python
destinationsDf = airportsDf.rename(columns = {
    'Origin': 'Destination',
    'Origin_Airport': 'Destination_Airport',
    'Origin_City': 'Destination_City',
    'Origin_Lat': 'Destination_Lat',
    'Origin_Long': 'Destination_Long'
})
destinationsDf.head()
```

```python
finalDf = pd.merge(totalDf, destinationsDf, on='Destination')
```

```python
finalDf
```

```python
cleanDf = finalDf[~(finalDf['Fare'] > 2000)]
```

```python
cleanDf = cleanDf[~(cleanDf['Fare'] < 50)]
cleanDf.head()
```

```python
cleanDf = cleanDf[~(cleanDf['Number_Passengers'] > 100)]
cleanDf.head()
```

```python
cleanDf.to_csv("flightData.csv", index=False)
```
